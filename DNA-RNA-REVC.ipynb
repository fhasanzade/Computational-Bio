### 3. The Secondary and Tertiary Structures of DNA

In <b><font color='green'>DNA</font></b> strings, symbols <font color='blue'>'A'</font> and <font color='blue'>'T'</font> are complements of each other, as are <font color='blue'>'C'</font> and <font color='blue'>'G'</font>.

The reverse complement of a <b><font color='blue'>DNA</font></b> string <I>s</I> is the string <I>s_c</I> formed by reversing the symbols of <I>s</I>, then taking the complement of each symbol (e.g., the reverse complement of <b><font color='red'>"GTCA"</font></b> is <b><font color='red'>"TGAC"</font></b>).

<b><font color='red'>Given</font></b>: A <b><font color='blue'>DNA</font></b> string <I>s</I> of length at most 1000 bp.

<b><font color='red'>Return</font></b>: The reverse complement <I>s_c</I> of <I>s</I>.

```python
dna3 = "CCAATACGATTGGTATAAGCCGATTTCTCACACTTGCAGGACCAAGATACATCAACTAGCTTGTCATCGCCTCTATCCGAAGAGAGGCATGGTTCCGTGGCCTCAACACTCCCACTGGGTAGTTCTGGGAACAATCCAGCGAATGTCAGTGGACAGTGACACCCGATGAGACGCCGCCCCGGTTGTTCTAGCCTCATTCCCTTGGATTTCCCAGCGCACAGTGAAATGTGCGTTGATTATGGTGCCCGACTTGAACTCCTTGATTAGTAATACACTAACAGATGAAAATTGCCAACTAGACAAACAGGAACGGTCGGAGAGCCCAGAATGGATCACTTCCACCTGGGTTGAATTGCCACACTCAGGTAAAGAAGTTTCGTCAGGGCGATCACGACGCTTAGGCAATAGTGCTTCCGCCCTACGAGCGGTCTGGCTGGAGCACCTGTTAGGAACTTACGCGCTAGTCAAATCCTACAAATAATTCGTTGCTTCTTGATATACGTCTAAGTCCGCAGGGAGTGCGGATCCTGGGCAAAATTAACTCTCCCCCTGAGATATTACCTGATGAGACCCTAGCAGCCTCCGTTTTGTTTGCTAAAACATAGTATAGCCCTGCTTGTACTGGAATCTTAGTGTTAGGGAACCGACAGTACATGAGCGTAACCATAGACTTACCGTCACTCTGGAATACGGCCAGAACTTCAACCTCACCGGACTAGGGCACAACGTAGGCGCGTCCCCGCTGGGGGCAAGTGTCCGTGTTCTTTCAAAAATGTGGACACATCACTGCACATTCCATCGATGTGGATGACAAACAGTAACCAGCGGACTTCTGAGGCGCCAAGCGTGCTTTTTGTCCCGTGCCTGGCC"
reverse=dna3[::-1]

s_comp=''
for n in reverse:
    if n == 'A':
        s_comp+='T'
    elif n == 'T':
        s_comp+='A'
    elif n == 'G':
        s_comp+='C'
    else:
        s_comp+='G'
        
print(s_comp)
```
